```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
df.drop('Id', axis = 1, inplace = True)
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size = 0.6)
```

```python
class KNN:
    
    def __init__(self, n_neighbours):
        self.n_neighbours = n_neighbours
        
    def euclidean_distance(self, x1, x2):
        return np.sum(np.square(x1 - x2)**2)
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    def _predict(self, x):
        distances = []
        
        for x_train in self.X_train:
            distance = self.euclidean_distance(x, x_train)
            distances.append(distance)
        distances = np.array(distances)
        
        n_neighbour_idxs = np.argsort(distances)[: self.n_neighbours]
        
        labels = y_train[n_neighbour_idxs]
        labels = list(labels)

        most_occuring_value = max(labels, key=labels.count)
        
        return most_occuring_value
    
    def predict(self, X):

        predictions = []
        
        for x in X:
            prediction = self._predict(x)
            predictions.append(prediction)
            
        return np.array(predictions)
```

```python
model = KNN(7)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)

print(f'The accuracy score of our model is {accuracy}')
```
