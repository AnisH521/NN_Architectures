```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/kaggle/input/iris/Iris.csv')
df.drop('Id', axis = 1, inplace = True)
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
class k_means:
    
    def __init__(self, K):
        self.K = K
        
    def initialize_centroids(self, X, K):    
        randomized_X = np.random.permutation(X.shape[0])

        centroid_indx = randomized_X[:K]
        centroids = X[centroid_indx]
    
        return centroids
    
    def assign_points_centroids(self, X, centroids):
        X = np.expand_dims(X, axis = 1)
        distance = np.linalg.norm((X - centroids), axis = -1)
        points = np.argmin(distance, axis = 1)
        
        return points
    
    def compute_mean(self, X, points, K):
        centroids = np.zeros((K, X.shape[1]))
        for i in range(K):
            centroid_mean = X[points == i].mean(axis = 0)
            centroids[i] = centroid_mean
            
        return centroids
    
    def predict(self, X, iterations = 10):
        centroids = self.initialize_centroids(X, self.K) 
        
        for _ in range(iterations):    
            points = self.assign_points_centroids(X, centroids)
            centroids = self.compute_mean(X, points, self.K)

        return centroids, points
```

```python
model = k_means(3)
y_pred = model.predict(X.values)
```

```python
label_encoder = LabelEncoder()
encoded_y = label_encoder.fit_transform(y)
```

```python
accuracy = accuracy_score(encoded_y, y_pred[1])
print(f'the accuracy of our model : {accuracy}')
```
